```python
from urllib.request import urlretrieve 
import pandas as pd

```

```python
url_of_file='https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
urlretrieve(url_of_file,'winequality-red.csv')
```

```python
df=pd.read_csv('winequality-red.csv',sep=';')
print(df.head())
```

```python
url_of_file='https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/'
sub_url_list=['winequality-red.csv','winequality-white.csv','winequality.names']
for url_list in sub_url_list:
    urlretrieve(url_of_file+url_list, str(url_list))
```

```python
import matplotlib.pyplot as plt

pd.DataFrame.hist(df.iloc[:,0:1])
plt.xlabel('fixed acidity (g(tartaric acid)/dm$^3$)')
plt.ylabel('count')
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup
url= 'https://www.python.org/~guido/'
r = requests.get(url)
html_doc = r.text
soup= BeautifulSoup(html_doc)
prettify_soup = soup.prettify()
print(prettify_soup)
```

```python
soup= BeautifulSoup(html_doc,"html.parser")
guido_title= soup.title
guido_text= soup.get_text()
a_tags= soup.find_all('a')
for link in a_tags:
    print(link.get('href'))
```
